```python
import time
import csv
from itertools import combinations    

def apriori(final_dict , min_support , min_confidence ):
    start_time = time.time()
    n_length = [len(i) for i in final_dict.values()]
    temp_items_lst = set()
    itemsets_count = list()   
    for i in final_dict.values():
        for j in i:
            temp_items_lst.add(j)
    item_list=()
    item_list = list(temp_items_lst)[:]
    item_counts = dict()  
    item_counts = item_frequency(temp_items_lst, final_dict)   
    items_min_support = dict()
    itemsets_count.append(item_counts)
    for key,value in item_counts.items():
        if (value/len(item_list))>min_support:
            items_min_support[key]= value   
    for i in range(2, max(n_length)+1):
        item_lst = combinations(temp_items_lst, i)
        item_counts = item_frequency(item_lst, final_dict, check=True)
        itemsets_count.append(item_counts)
        for key,value in item_counts.items():
            if (value/len(item_list))>min_support:
                items_min_support[key]= value
    print("\nThe itemsets satisfying the" + "\033[1m" + " minimum support criterion" +"\033[0m" + " are:")
    print("\033[1m" + "Itemset\tSupport" + "\033[0m")
    for key, value in items_min_support.items():
        print(key, ' : ', round((value/len(item_list))*100,3) , "%")
    association_minsupport = min_support_itemset_association(items_min_support)
    print("\n The corresponding association rules are:")
   
    ans = {}
    for i in association_minsupport:
        for j in i.items():
            if type(j[0]) == type(str()):
                independent = {j[0]}
            else:
                independent = set(j[0])
            if type(j[1]) == type(str()):
                dependent = {j[1]}
            else:
                dependent = set(j[1])
            for m in items_min_support:
                if type(m) != type(str()):
                    if independent.union(dependent) - set(m) == set():
                        numerator_conf = items_min_support[m]
                    if len(dependent) == len(set(m)) and dependent - set(m) == set():
                        denominator_conf = items_min_support[m]
                else:
                    if len(dependent) >= len({m}):
                        if dependent - {m} == set():
                            denominator_conf = items_min_support[m]
                    elif len(dependent) <= len({m}):
                        if {m} - dependent == set():
                            denominator_conf = items_min_support[m]
   
            if numerator_conf/denominator_conf >= min_confidence:
                ans[tuple(independent)[0]] = dependent, numerator_conf/denominator_conf , j[1]
                confidence=round((numerator_conf/denominator_conf )*100,2)
            
                output = "{} => {} \t Confidence:{}% \n"
                print(output.format(dependent, independent, confidence))
```

```python
def min_support_itemset_association(items_min_support):
    items_association= []
    for i in items_min_support:
        association = {}
        if type(i) != type(str()):
            i = list(i)
            temp_i = i[:]
            for j in range(len(i)):
                k = temp_i[j]
                del temp_i[j]
                association[k] = temp_i
                temp_i = i[:]
        items_association.append(association)
    items_association = list(filter(None, items_association))
    return items_association
```

```python
def item_frequency(item_list, final_dict , check=False):
            item_associations = dict()
            for i in item_list:
                temp_i = {i}
                if check:
                    temp_i = set(i)
                for j in final_dict.items():
                    if temp_i.issubset(set(j[1])):
                        if i in item_associations:
                            item_associations[i] += 1
                        else:
                            item_associations[i] = 1
            return item_associations 
```

```python
def conv_dict(file):
    output = dict()
    with open(file) as f:
        filedata = csv.reader(f, delimiter=',')
        count = 0
        for line in filedata:
            count += 1
            output[count] = list(set(line))
    return output
```
